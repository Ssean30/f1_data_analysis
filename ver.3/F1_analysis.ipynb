```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X = df['grid'].values.reshape(-1,1)
y = df['positionOrder'].values.reshape(-1,1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(StandardScaler(), PCA(n_components=1),LogisticRegression(random_state=1, solver='lbfgs') )
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
print('predictions %s' % y_pred)
print('Accuracy %.3f' % pipeline.score(X_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(estimator=pipeline, X=X_train, y=y_train, cv=10, n_jobs=1)
print('%s' % scores)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```
